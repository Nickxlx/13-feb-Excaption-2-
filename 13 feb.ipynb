Q1. Explain why we have to use the Exception class while creating a Custom Exception.

Answer--> When creating a custom exception in Python, it is recommended to subclass the built-in Exception class. This is because the Exception class provides a basic structure for an exception and includes useful methods and attributes and By subclassing Exception, the custom exception inherits these methods and attributes and can also be caught by generic exception handlers that catch all Exception subclasses.

Q2. Write a python program to print Python Exception Hierarchy.

```python
# print the Python Exception Hierarchy
def print_exception_hierarchy():
    for exc in Exception.__subclasses__():
        print(exc.__name__)
        for sub_exc in exc.__subclasses__():
            print("  ", sub_exc.__name__)
            for sub_sub_exc in sub_exc.__subclasses__():
                print("    ", sub_sub_exc.__name__)
            if len(sub_exc.__subclasses__()) == 0:
                print("    No more subclasses")

print_exception_hierarchy()
```

Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

Answer--> LookupError is a subclass of the Exception class, and it has several subclasses that are more specific to different types of lookup or indexing errors.
              Two common subclasses of LookupError are KeyError and IndexError.

KeyError is raised when trying to access a non-existent key in a dictionary.

```python
# For example:

d = {"a":[1,2,3,45], "b":365}

print(d["c"]) # this will gives a KeyError
```

IndexError is raised when trying to access an index that is out of range in a sequence like a list or a tuple.

```python
# For example:
l = [1,23,4,56,7,7]

print(l[7]) # this will give a IndexError
```

Q5. Explain ImportError. What is ModuleNotFoundError?

Answer--> ImportError is a general exception that is raised when there is a problem importing a module. It can occur in several situations, such as when the module you are trying to import does not exist, or when there is a syntax error in the module.

On the other hand, ModuleNotFoundError is a more specific exception that is raised when a module cannot be found during the import process. It is a subclass of ImportError and was introduced in Python 3.6.
